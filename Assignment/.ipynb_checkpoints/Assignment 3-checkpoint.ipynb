## T3. Missing values

```python
airquality.isna().sum() #  missing values in column of data
```

```python
airquality.isna().sum().sum() #total missing values
```

```python
airquality1 = airquality.copy()
```

```python
airquality1.head()
```

```python
airquality1 = airquality1.fillna(airquality1.mean())
```

```python
airquality1.isna().sum()
```

```python
airquality1.fillna({"Ozone" : airquality1["Ozone"].median(), "Solar.R" : airquality1["Solar.R"].mean()})
```

###  Why might median be preferred over mean in some cases?
- If the data is skewed (not symmetric), if datasets consists of extreme values median is a better measures and more reliable.
- For filling missing values (NaN),median prevents bias caused by a few extreme numbers.
* Conclusion:From the "Airquality" dataset columns like Ozone and Solar.R often have extreme values (very high readings on some days).\
    So, Its better to use median instead mean

## T5. Filtering & conditions

```python
filtered_airquality = airquality[(airquality['Ozone'] > 100) | (airquality['Temp'] >= 85)]
filtered_airquality
```

```python
#Filter rows where Solar.R is missing.
missing_Solar = airquality[airquality['Solar.R'].isnull()]
print(missing_Solar)
```

```python
#Filter rows for July (7) with Ozone between 50 and 100 (inclusive).
filtered_inclusive = airquality[(airquality["Month"]== 7) & (airquality["Ozone"].between(50 , 100))]
filtered_inclusive
```

## T6. Sorting

```python
#Sort the dataset by Ozone descending and Temp ascending (multi-column sort).
multi_column_sorted_airquality = airquality.sort_values(by=["Ozone","Temp"],ascending=[False,True])
multi_column_sorted_airquality
```

```python
#Sort the dataset by Month then Day to get chronological order
MDorder_airquality =airquality.sort_values(by=["Month","Day"],ascending=[True,True])
MDorder_airquality
```

# T7 Duplicates

```python
#Check if there are any fully duplicated rows
airquality.duplicated().sum()
```

```python
#Intentionally create a few duplicate rows (e.g., duplicate first 3 rows) into df_dup.
df_dup = pd.concat([airquality, airquality.iloc[0:3]], ignore_index=True)
df_dup.tail()
```

```python
#Remove duplicates and report how many were removed.
df_rem_dup =airquality.drop_duplicates()
df_rem_dup.tail()
```

```python
removed_count = len(df_dup) - len(df_rem_dup)
removed_count
```
