# simulate duplicate rows

```python
mtcars2[mtcars2["brand_car"] == "Mazda RX4"] #all brand_car with name of Mazda RX4
```

```python
mazda2 = mtcars2[mtcars2["brand_car"] == "Mazda RX4"]
datson2 = mtcars2[mtcars2["brand_car"]== "Datsun 710"]
# copied mazda and datson car in variable/ row 
```

```python
mazda2
datson2
```

```python
mtcars_dup = pd.concat([mtcars2,mazda2,datson2,mazda2,datson2], ignore_index=True) # 2 more cars Mazda, 2 more car for Datson
mtcars_dup #.sort_values('car')
mtcars_dup
```

```python
mtcars_dup.shape
```

```python
mtcars_dup[mtcars_dup["brand_car"]== "Datsun 710"] #only shows Datsun 710" car from the data
```

```python
mtcars_dup[mtcars_dup["brand_car"] == "Mazda RX4"]
```

```python
mtcars_dup[mtcars_dup["brand_car"] == "Volvo 142E"]
```

```python
mtcars_dup.duplicated().sum()
```

```python
mtcars_dup.duplicated(subset = ["gear"]).sum()
```

```python
mtcars_dup.duplicated(subset = ["disp"]).sum()
```

```python
mtcars_dup.duplicated(subset = ["disp","gear","hp"]).sum() #every values of the columns must be same to be dup ie sum 8
```

```python
#mtcars_dup.duplicated(keep=False)  #mark all duplicate
```

```python
mtcars_dup.drop_duplicates(keep="last") #last priority or can keep "first"
```

```python
mtcars_dup.drop_duplicates(keep="first")
```

```python
mtcars_dup.drop_duplicates(keep = False) #drop all duplicates
```

```python
mtcars[mtcars.duplicated(subset = ["gear","am","cyl","carb"],keep=False)].sort_values(by=["gear","am","cyl","carb"], ascending=[True, False, True, False])
#ascending = True is by default, optional, or can skip
#keep=False marks all occurrences of duplicates, not just the later ones.
```

```python
mtcars_dup[~ mtcars_dup.duplicated(subset = ["gear","cyl","am","carb"],keep=False)].sort_values(by=["gear","cyl","am","carb"])
# non duplicates with respect to column in subset
```

```python
mtcars_dup.drop_duplicates(subset =["gear","cyl","am","carb"],inplace=False) #remove duplicates by selected columns
```
