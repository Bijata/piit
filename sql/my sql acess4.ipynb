## Method 1: Using Python mysql library

```python
# libraries
import pandas as pd
import pymysql
from sqlalchemy import create_engine , text
import os
```

### 1. Connect juypter with mysql to work on mysql databases with create engine

```python
#engine = create_engine('mysql+pymysql://piuser:Pass#123@localhost:3306/piit') (mandatory to create engine to read from sql)
engine1 = create_engine("mysql+pymysql://piuser:Pass%40123@localhost:3306/piit")
print(engine1)
```

### 2. Read from mysql

```python
pd.read_sql('SHOW DATABASES;', engine1)
```

```python
pd.read_sql('SHOW TABLES;', engine1)
```

```python
pd.read_sql('SELECT * FROM empdata2;', engine1)
```

```python
pd.read_sql('SELECT * FROM empdata2 ;', engine1)
```

```python
pd.read_sql('SELECT * FROM empdata2 WHERE salary > 50000.0;', engine1)
```

```python
pd.read_sql('SELECT * FROM empdata2 ORDER BY employee_id DESC;', engine1)
```

```python
pd.read_sql('SELECT MIN(salary) FROM empdata2;', engine1)
```

```python
pd.read_sql('SELECT Avg(salary) FROM empdata2;', engine1)
```

```python
pd.read_sql('SELECT COUNT(*) FROM empdata2;', engine1)
```

```python
pd.read_sql('SELECT * FROM empdata2 WHERE salary BETWEEN 60000.0 AND 70000.0  ;', engine1)
```

```python
# function to run the query with parameters or query (UPDATE, DELETE CAN B USE BY PARAM)
def q(sql, **params):
    with engine1.connect() as conn:
        return pd.read_sql(text(sql), conn, params=params)
```

```python
q('SELECT * FROM empdata2 WHERE first_name = "Jane"')
```

```python
# Define SQL query function
pysqldf = lambda q: sqldf(q, globals())
```

```python
pysqldf("SELECT * FROM mtcars2 where mpg >25;")
```

```python
pysqldf("SELECT * FROM mtcars2 where mpg >25 AND vs = 1;")
```
